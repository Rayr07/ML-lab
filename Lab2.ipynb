```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries 'matplotlib.pyplot' and 'seaborn' imported successfully.")

mean_total_sulfur_dioxide = df.groupby('fixed_acidity')['total_sulfur_dioxide'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='fixed_acidity', y='total_sulfur_dioxide', data=mean_total_sulfur_dioxide)
plt.title('Mean Total Sulfur Dioxide vs. Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Mean Total Sulfur Dioxide')
plt.grid(True)
plt.show()

quality_counts = df['quality'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=quality_counts.index, y=quality_counts.values, hue=quality_counts.index, palette='viridis', legend=False)
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality Rating')
plt.ylabel('Count')
plt.show()
```
