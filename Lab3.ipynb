```python
import pandas as pd
```

```python
file_path = '/content/StudentsPerformance.csv'
df = pd.read_csv(file_path)
print("Displaying first few records of dataset:")
df.head()
```

```python
print("Dataset Info:")
df.info()
```

```python
print("Dataset dtypes:")
df.dtypes
```

```python
print("Measures of Central Tendency of Dataset:")
df.describe()
```

```python
numerical_cols=df.loc[:,['math score','reading score','writing score']]
print(" Quartile Analysis: ")
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q2 = df[col].quantile(0.50)
    q3 = df[col].quantile(0.75)

    print(f"\nStatistics for '{col}':")
    print(f"Q1 (25th percentile): {q1:.2f}")
    print(f"Q2 (50th percentile/Median): {q2:.2f}")
    print(f"Q3 (75th percentile): {q3:.2f}")
```

```python
numerical_cols = ['math score', 'reading score', 'writing score']
print("Correlation Matrix")
correlation_matrix = df[numerical_cols].corr()
print(correlation_matrix)
print("Covariance Matrix")
covariance_matrix = df[numerical_cols].cov()
print(covariance_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_cols = ['math score', 'reading score', 'writing score']
plt.figure(figsize=(18, 5))
plt.suptitle('Histograms of Numerical Scores', fontsize=16)
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f'{col.replace("_", " ").title()} Distribution')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(18, 5))
plt.suptitle('Boxplots of Numerical Scores', fontsize=16)
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'{col.replace("_", " ").title()} Outliers')
    plt.ylabel('Score')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
