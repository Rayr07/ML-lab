```python
import numpy as np
a = [1, 2, 3, 4, 5]
print("Original List:",a)
np_arr = np.array(a)
print("NumPy Array:",a)
print("Shape of the array:", np_arr.shape)
print("Data type of the array:", np_arr.dtype)
```

```python
import numpy as np
z = np.zeros((3, 4))
print(f"Zeros Matrix:\n{z}")
print(f"Shape of Matrix: {z.shape}\n")

o = np.ones((2, 5))
print(f"Ones Matrix:\n{o}")
print(f"Shape of Matrix: {o}")
```

```python
import numpy as np
a = np.arange(1, 13)
print("Original Array:",a)
print("Shape of Array:",a)
r = a.reshape(3, 4)
print(f"Reshaped Array:\n {r}")
print(f"Shape of Reshaped Array:\n {r}")
```

```python
import numpy as np

a = np.random.randint(1, 101, 10)
print("Random Integers Array:",a)

print(f"Minimum value:", a.min())
print(f"Maximum value:",a.max())
print(f"Mean value:", a.mean())
print(f"Sum of values:", a.sum())
```
